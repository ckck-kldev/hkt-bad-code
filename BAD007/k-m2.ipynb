```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate sample data
X, y = make_blobs(n_samples=5000, n_features=2, centers=15, shuffle=True, random_state=np.random.randint(10))

# Run KMeans algorithm
NUM_CLUSTERS = 5
km = KMeans(n_clusters=NUM_CLUSTERS, init='random', n_init=10, max_iter=400, tol=1e-04, random_state=2)
y_km = km.fit_predict(X)

# Plot data points with different colors and shapes for each cluster
plt.figure(figsize=(8, 6))
colors = ['b', 'g', 'r', 'c', 'm']
markers = ['o', 's', 'v', '^', 'p']
for i in range(NUM_CLUSTERS):
    plt.scatter(X[y_km == i, 0], X[y_km == i, 1], c=colors[i], marker=markers[i], label=f'Cluster {i}')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=250, marker='*', c='red', label='Centroids')
plt.legend(loc='best')
plt.grid()
plt.show()
```
